```python
labels = ['A' , 'B ' , 'C']
data = [80 , 40 , 60]
explode = (0.1 , 0.1, 0.1 )
custom_color = ('#099098' , '#009988' , '#4DAF32')

plt.pie(x = data , labels =labels , explode = explode , colors = custom_color , autopct='%1.1f%%')
plt.show()
```

```python
data2 = (50 , 70 , 30)

plt.bar(labels , data)
plt.bar(labels , data2)
plt.show()
```

```python
plt.barh(labels , data , color = custom_color)
plt.show()
plt.savefig('bar plot.png')
```
