```python
mapper = {'Spring' : 1 ,
          'Summer' : 3 ,
          'Winter' : 0 ,
          'Fall' : 2}
df['season'] = df['season'].map(mapper)
```

```python
from collections import Counter
```

```python
Counter(df['weather'])
```

```python
df['weather'] = df['weather'].replace('Heavy rain/snow' , 'Light rain/snow')
```

```python
Counter(df['weather'])
```

```python
mapper = {'Clear' : 0 ,
          'Cloudy' : 1 ,
          'Light rain/snow' : 2 ,
        }
df['weather'] = df['weather'].map(mapper)
```

```python
Counter(df['holiday'])
```

```python
mapper = {'Yes' : 0,
         'No' : 1,
         }

df['holiday'] = df['holiday'].map(mapper)
```

```python
df.sample(10)
```

```python
df.drop_duplicates(inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
to_scale = ['datetime' , 'temp' , 'humidity' , 'windspeed']

for each_feature in to_scale:
    scaler = StandardScaler()
    df[each_feature] = scaler.fit_transform(df[each_feature].values.reshape(-1 , 1))
```
