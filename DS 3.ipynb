```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
meu = np.array([3 , 2]) #3 is mean for x-aixs and 2 is for y-axis
sigma = np.array([
    [10  ,0], # 10 is std for x-axis
    [0 , 1] # 1 is std for y-axis
])
```

```python
multi_norm = np.random.multivariate_normal(meu , sigma , 1000)
```

```python
multi_norm.shape
```

```python
multi_norm[:5]
```

```python
x = multi_norm[: , 0] # multi_norm [row_slicing , column_slicing , step]
y = multi_norm[: , 1]
```

```python
plt.scatter(x , y)
plt.title('Title')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.hist2d(x , y)
plt.show()
```

```python
np.corrcoef(x , y)
```
